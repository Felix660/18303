```julia
using PyPlot
```

```julia
# construct the (M+1)xM matrix D, not including the 1/dx factor
diff1(M) = [ [1.0 zeros(1,M-1)]; diagm(ones(M-1),1) - eye(M) ]
```

```julia
# finite-difference (FD) Green's function:
N=100
L = 1
dx = L/(N+1)
D = diff1(N)/dx
function A(ω)
  return D'*D - ω^2*eye(N)
end
function d(x)
  k = iround(x/L * (N+1))
  k = k < 1 ? 1 : (k > N ? N : k)
  dk = zeros(N)
  dk[k] = 1/dx
  return dk
end
g(x,ω) = A(ω) \ d(x)

# analytical Green's function:
G(x,y,ω) = (x < y ? sin(ω*x)*sin(ω*(L-y)) : sin(ω*y)*sin(ω*(L-x))) / (ω * sin(ω*L))
G(X::Array,y,ω) = [ G(x,y,ω) for x in X ]
```

```julia
x1 = [1:N] * dx
x2 = linspace(0,L,200)

subplot(1,2,1)
ω = 0.4*pi/L
plot(x2, G(x2, 0.5, ω), "b-")
plot(x1, g(0.5, ω), "bo")
plot(x2, G(x2, 0.75, ω), "r-")
plot(x1, g(0.75, ω), "ro")
xlabel(L"$x/L$")
title(L"frequency $\omega = 0.4 \pi / L$")
legend([L"exact $x'=0.5L$", L"FD $x'=0.5L$", L"exact $x'=0.75L$", L"FD $x'=0.75L$"],"lower right")

subplot(1,2,2)
ω = 5.4*pi/L
plot(x2, G(x2, 0.5, ω), "b-")
plot(x1, g(0.5, ω), "bo")
plot(x2, G(x2, 0.75, ω), "r-")
plot(x1, g(0.75, ω), "ro")
xlabel(L"$x/L$")
title(L"frequency $\omega = 5.4 \pi / L$")

savefig("pset6-f13-2d.pdf")
```

Hooray, we have a pretty good match between the exact $G(x,x')$ Green's function and the finite-difference (FD) approximation.  At the higher frequency, the match is not quite as good, but that isn't too surprising: -the difference approximations become less and less accurate as the function becomes more and more oscillatory relative to the grid $\Delta x$.
