```julia
using PyPlot
using PyCall
@pyimport scipy.optimize as s

# First plot the left- and right-hand sides of the equation kL = tan(kL) to solve
kL = linspace(0,6pi,1000)
plot(kL/pi, kL, "k--")
plot(kL/pi, tan(kL), "r-")
axis([0,6, 0,20])
xlabel(L"$kL/\pi$")

# Use Newton's method to solve tan(x)-x for statring points close to
# 3pi/2, 5pi/2, ..., 11pi/2.
kLn = [ s.newton(x -> tan(x)-x, n*pi/2-0.1) for n in 3:2:11 ]

# plot the solutions as dots, add a legend, and save to PDF
scatter(kLn/pi, kLn)
legend([L"$kL$", L"$\tan(kL)$", "solutions"], "upper left")
savefig("pset1-f13-p2ai.pdf")
```

```julia
# The eigenvalues are -k^2:
print(-kLn.^2)
```
