## part (c)

In this part, we compute a center-difference approximation of $(cu')'$ for $c=e^{3x}$ and $u=\sin(x)$, and compare it to the exact answer at $x=1$ as a function of the discretization spacing $\Delta x$.  We expect second-order convergence.

```julia
using PyPlot
x=1
dx = logspace(-8,-1,50)
c(x) = exp(3x)
u(x) = sin(x)
approx = (c(x+dx/2) .* (u(x+dx)-u(x))./dx - c(x-dx/2) .* (u(x)-u(x-dx))./dx) ./ dx
exact = 3*exp(3x) * cos(x) - exp(3x) * sin(x) # = c'u' + cu''
loglog(dx, abs(approx - exact), "bo")
loglog(dx, dx.^2, "k-")
legend(["|error|", L"$\Delta x^2$"], "lower right")
xlabel(L"$\Delta x$")
ylabel("|error| in derivative")
savefig("pset2-f13-p1c.pdf")
```

Notice two things from this plot.  First, the error points decrease at the expected $\sim \Delta x^2$ rate, since they are parallel to the $\Delta x^2$ line on the log–log scale.  Second, once $\Delta x$ gets $\lesssim 10^{-4}$, the errors *stop* decreasing, because they become limited by roundoff errors in the computer's arithmetic (floating-point arithmetic, $\sim 15$ significant digits), and in fact the errors actually start *increasing*.  The problem is that the roundoff errors are getting bigger and bigger for small $\Delta x$ because our formula subtracts nearly equal quantities, an example of "catastrophic cancellation" errors.  In solving PDEs, however, this is rarely a practical issue because we almost never take $\Delta x$ to be so small.

## part (f)

```julia
# construct the (M+1)xM matrix D, not including the 1/dx factor
diff1(M) = [ [1.0 zeros(1,M-1)]; diagm(ones(M-1),1) - eye(M) ]
```

```julia
# we will make this a function of M so that we can easily vary M to check convergence below:
function computeAeig(M)
    L = 1
    D = diff1(M)
    dx = L / (M+1)
    x = dx*0.5:dx:L # sequence of x values from 0.5*dx to <= L in steps of dx
    C = diagm(exp(3*x)) # diagonal matrix of c(x) values at the points x
    A = -D' * C * D / dx^2
    return eig(A)
end
λ, U = computeAeig(100)
```

```julia
# As expected, since A is negative-definite, all its eigenvalues are < 0
all(λ .< 0)
```

```julia
using PyPlot
plot(dx:dx:L-dx, U[:,end-3:end])
xlabel("x"); ylabel("eigenfunctions")
legend(["fourth", "third", "second", "first"])
savefig("pset2-f13-p1f.pdf");
```

```julia
# verify orthogonality (up to roundoff errors)
dot(U[:,end], U[:,end-1])
```

For the last part, you are asked to check that eigenvalue differences $|\lambda_{2M} - \lambda_{M}|$ decrease $\sim 1/M^2$.  You can do this just by looking at the numbers, and by printing out the ratios we see that doubling $M$ decreases this difference by almost exactly a factor of 4, but for the solutions it is a bit prettier to show a plot: 

```julia
Ms = [100,200,400,800,1600]
λs = map(M -> computeAeig(M)[1][end], Ms) # compute λ[end] for M in [100,200,400,800]
for i = 2:length(Ms)-1
    println("|λ($(Ms[i])) - λ($(Ms[i-1]))| / |λ($(Ms[i+1])) - λ($(Ms[i]))| = ", abs(λs[i]-λs[i-1]) / abs(λs[i+1]-λs[i]))
end
loglog(Ms[1:end-1], abs(diff(λs)), "r*-")
loglog(Ms/1.2, 200./Ms.^2, "k-")
xlabel(L"number of grid points $M$")
legend([L"$|\lambda_{2M} - \lambda_{M}|$", L"$\sim 1/M^2$"]);
```

# Problem 2

First, let's repeate the calculations from pset 1 in order to find the first two eigenvalues $\lambda = -k^2$ by solving for $kL$ from the transcendental equation $\tan(kL) = kL$.

```julia
using PyCall
@pyimport scipy.optimize as s
kL1 = s.newton(kL -> tan(kL) - kL, 3pi/2 - 0.1)
kL2 = s.newton(kL -> tan(kL) - kL, 5pi/2 - 0.1)
(kL1/pi,kL2/pi)
```

Let's plot them, as in pset 1, to remind ourselves of what they look like:

```julia
# define the eigenfunctions:
L = 1
u1(x) = sin(kL1*x/L)
u2(x) = sin(kL2*x/L)

# plot them
x = linspace(0,L,200)
plot(x, u1(x))
plot(x, u2(x))
legend([L"$u_1$", L"$u_2$"], "lower left");
```

Finally, we'll use the `quadgk` function in Julia, which performs integrals numerically, to compute $\int_0^L u_1(x) u_2(x) dx$.  The `abstol=1e-13` parameter indicates that we want to compute the integral to at least 13 decimal places if possible.   The `quadgk` function returns a pair $(I,E)$ of values, where $I$ is the estimated integral and $E$ is the estimated error.

```julia
quadgk(x -> conj(u1(x)) * u2(x), 0, L, abstol=1e-13)
```

As desired, the estimated integral $-7.5 \times 10^{-16}$ is essentially zero up to round-off errors.
